# Estudiar como hacer bien la tabla y los tipos de tablas que se pueden hacer

```python

# Filtrar las canciones más populares (opcional: limitar a las 10 más populares)
top_populares = df_Rock.sort_values(by='popularity', ascending=False).head(10)

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))  # Tamaño de la gráfica
plt.barh(top_populares['name'], top_populares['popularity'], color='skyblue')  # Gráfica horizontal

# Personalizar la gráfica
plt.xlabel('Popularidad', fontsize=12)
plt.ylabel('Canciones', fontsize=12)
plt.title('Canciones más populares', fontsize=16)
plt.gca().invert_yaxis()  # Invertir el orden para que la más popular esté arriba
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
