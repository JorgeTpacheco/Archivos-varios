# Estudiar como hacer bien la tabla y los tipos de tablas que se pueden hacer

```python

# Filtrar las canciones ms populares (opcional: limitar a las 10 ms populares)
top_populares = df_Rock.sort_values(by='popularity', ascending=False).head(10)

# Crear la grfica de barras
plt.figure(figsize=(10, 6))  # Tamao de la grfica
plt.barh(top_populares['name'], top_populares['popularity'], color='skyblue')  # Grfica horizontal

# Personalizar la grfica
plt.xlabel('Popularidad', fontsize=12)
plt.ylabel('Canciones', fontsize=12)
plt.title('Canciones ms populares', fontsize=16)
plt.gca().invert_yaxis()  # Invertir el orden para que la ms popular est arriba
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Mostrar la grfica
plt.tight_layout()
plt.show()
```
