* Obten el numero de veces que se manda una campaÃ±a a los clientes, y has un histograma o grafica de barras en el que se pueda observar en el eje x, las veces que se manda una campaÃ±a y en el eje y, la cantidad de clientes que cae en cada grupo.


```python
df_campaÃ±as.columns
```

```python
conteo = df_campaÃ±as.groupby(['IdSubscriber', 'Campaign']).size().reset_index(name='veces_enviada')
conteo.head(5)
```

```python
clientes_por_grupo = conteo['veces_enviada'].value_counts().sort_index()
```

```python
plt.figure(figsize=(8, 5))
clientes_por_grupo.plot(kind='bar', color='Red', edgecolor='black')

# Personalizar grÃ¡fico
plt.title('DistribuciÃ³n de envÃ­os por cliente', fontsize=14)
plt.xlabel('Veces que se enviÃ³ la campaÃ±a', fontsize=12)
plt.ylabel('Cantidad de clientes', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar grÃ¡fico
plt.tight_layout()
plt.show()
```
